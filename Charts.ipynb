```python
import pandas as pd
df= pd.read_excel('mean median mode.xlsx')
```

```python
df.plot.scatter(x='age',y='salary_2018')
```

```python
df1 = df[['age','salary_2018','percent_hike']]
```

```python
from pandas.plotting import scatter_matrix

scatter_matrix(df1, figsize=(12,12))
```

```python
df.cov()
```
